```python
import pandas as pd
import numpy as np
import xarray as xr
import cftime
import datetime
from natsort import os_sorted
from os.path import basename
from tqdm import tqdm
```

```python
import glob
fl=os_sorted(glob.glob('/content/drive/MyDrive/Colab Notebooks/TA/Data GSMap/CH/2003*.nc'))
fl
```

```python
g_ds=[]
for f in tqdm(fl):
  ds=xr.open_dataset(f)
  c=basename(f)
  ds['time']=pd.date_range(start=f'{c[0:4]}-{c[4:6]}-01',periods=ds.time.size,freq='H')
  g_ds.append(ds)
g_ds=xr.concat(g_ds,dim='time')
```

```python
g_ds.to_netcdf('/content/drive/MyDrive/Colab Notebooks/TA/Data GSMap/CH new/2003.nc')
```
