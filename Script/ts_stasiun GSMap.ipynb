```python
tabeltahun = []

for tahun in range(2000,2022):
  ds = xr.open_dataarray('/content/drive/MyDrive/Colab Notebooks/TA/SIDANG/DATA/CH GSMaP/{}.nc'.format(tahun))

  ikoordinat = []
  for i in range(len(data_centroid)):
    ikoordinat.append(
        ds.sel(lat=data_centroid.loc[i,'Latitude'],
              lon=data_centroid.loc[i,'Longitude'],
              method='nearest'
              ).to_pandas().rename('centroid_{}'.format(i))
    )
  ds.close()
  tabeltahun.append(pd.concat(ikoordinat, axis=1))

df_alltime_allstat = pd.concat(tabeltahun, axis=0)
df_alltime_allstat
```

```python
# Menyimpan data
df_alltime_allstat.to_csv('/content/drive/MyDrive/Colab Notebooks/TA/SIDANG/DATA/ts_stasiun_gsmap.csv', index=True)
```
